## Customizing plots

```python
import pandas as pd
df2 = pd.read_csv('./data/visualization/df2.csv')
df2.head()
```

```python
df2.plot.line()
```

```python
df2['c'].plot.line()
```

```python
df2['c'].plot.line(figsize=(10,3), ls=":", c="red")
```

```python
title = "MY PLOT TITLE"
xlabel = "My X data"
ylabel = "My y data"
```

```python
ax = df2['c'].plot.line(figsize=(10,3), ls=":", c="red", title=title)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

```python
ax = df2.plot()
ax.legend(loc=0)
```

```python
ax = df2.plot()
ax.legend(loc=0, bbox_to_anchor=(1.0,1.0))
```
